```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table'
```

```python
page = requests.get(url)
print(page.status_code)
```

```python
soup = bs(page.text, 'html')
print(soup.prettify())
```

```python
table = soup.find_all('table', class_ = 'wikitable sortable sticky-header-multi')[0]
print(table)
```

```python
table_head = table.find_all('tr')[1]
print(table_head)
```

```python
table_head_span = table_head.find_all('span')
print(table_head_span)
```

```python
headers = [title.get('title') for title in table_head_span if title.get('title')]
headers
```

```python
rows = table.find_all('tr')
rows
len(rows)
```

```python
data = pd.DataFrame(columns=headers)
data.head()
```

```python
table_data = []
for row in rows:
    row_data = row.find_all('td')
    row_data_parsed = [td.text.replace('\xa0', '').strip() for td in row_data]
    
    if len(row_data_parsed) > 0: data.loc[len(data)] = row_data_parsed
```

```python
data.head()
```

```python
data['International Olympic Committee country code'] = data['International Olympic Committee country code'].map(lambda x: x[:x.index(')') + 1])
```

```python
data.head()
```

```python
data['Country name'] = data['International Olympic Committee country code'].map(lambda x: x[:-5])
data['Country code'] = data['International Olympic Committee country code'].map(lambda x: x[-4:-1])
data.head()
```

```python
data = data.rename(columns={
    'International Olympic Committee country code': 'Country',
    'Number of Summer Games each country has participated in' : 'Summer games participated',
    'Number of Summer Gold Medals won by each country: No': 'Gold medals won (summer games)',
    'Number of Summer Silver Medals won by each country': 'Silver medals won (summer games)',
    'Number of Summer Bronze Medals won by each country': 'Bronze medals won (summer games)',
    'Total number of Summer Medals won by each country': 'Total medals won (summer games)',
    'Number of Winter Games each country has participated in' : 'Winter games participated',
    'Number of Winter Gold Medals won by each country': 'Gold medals won (winter games)',
    'Number of Winter Silver Medals won by each country': 'Silver medals won (winter games)',
    'Number of Winter Bronze Medals won by each country': 'Bronze medals won (winter games)',
    'Total number of Winter Medals won by each country': 'Total medals won (winter games)',
    'Combined total number of Games each country has participated in' : 'Total games participated',
    'Combined total number of Gold Medals won by each country' : 'Gold medals won (all games)',
    'Combined total number of Silver Medals won by each country' : 'Silver medals won (all games)',
    'Combined total number of Bronze Medals won by each country' : 'Bronze medals won (all games)',
    'Combined total number of all Medals won by each country' : 'All medals won (all games)'    
})
```

```python
data = data[['Country', 
             'Country name', 
             'Country code',
             'Summer games participated',
             'Gold medals won (summer games)',
             'Silver medals won (summer games)',
             'Bronze medals won (summer games)',
             'Total medals won (summer games)',
             'Winter games participated',
             'Gold medals won (winter games)',
             'Silver medals won (winter games)',
             'Bronze medals won (winter games)',
             'Total medals won (winter games)',
             'Total games participated',
             'Gold medals won (all games)',
             'Silver medals won (all games)',
             'Bronze medals won (all games)',
             'All medals won (all games)',
             ] ]
```

```python
data.head()
```

```python
num_cols = [col for i, col in enumerate(data.columns) if i>2]
num_cols
```

```python
data[num_cols] = data[num_cols].apply(lambda x: x.astype(int))
```
